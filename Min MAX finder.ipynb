```python
import pandas as pd

df = pd.read_csv('/content/ayush stray vacancy Round 3 closing air.csv')

def assign_course(code_college):
    if str(code_college).startswith('3'):
        return 'BAMS'
    elif str(code_college).startswith('4'):
        return 'BHMS'
    elif str(code_college).startswith('5'):
        return 'BUMS'


df['Course'] = df['code_college'].apply(assign_course)


df.to_csv('Ayush stray vacancy R-2 file.csv', index=False)

print("Course column added based on code_college values.")
```

```python
import pandas as pd

df = pd.read_csv('/content/Allied stray vacancy Round 2 closing air.csv')

def assign_course(code_college):
    if str(code_college).startswith('6'):
        return 'BPTH'
    elif str(code_college).startswith('8'):
        return 'BASLP'
    elif str(code_college).startswith('7'):
        return 'BOTH'
    elif str(code_college).startswith('9'):
        return 'BP&O'


df['Course'] = df['code college'].apply(assign_course)


df.to_csv('BPTH stray r-2 extracted file with course.csv', index=False)

print("Course column added based on code_college values.")
```
